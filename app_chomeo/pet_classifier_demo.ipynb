```python
import os
from pathlib import Path

def setup_path(folder_name):
    base_path = Path('/content/drive/MyDrive')
    target = base_path / folder_name

    if target.exists():
        global PROJECT_DIR, APP_DIR, APP_PATH, CATDOG_MODEL_PATH, BREED_MODEL_PATH
        PROJECT_DIR = target
        APP_DIR = PROJECT_DIR / 'app_chomeo'
        APP_PATH = APP_DIR / 'app_chomeo.py'
        CATDOG_MODEL_PATH = PROJECT_DIR / 'output' / 'main2' / 'best_cnn_cats_dogs_main2.keras'
        BREED_MODEL_PATH = PROJECT_DIR / 'output' / 'oxford_pet_mobilenetv2' / 'best_mobilenetv2_oxford_pet.keras'

        print(f'✅ Đã kết nối thành công: {PROJECT_DIR}')
        print(f'Kiểm tra thư mục APP: {"OK" if APP_DIR.exists() else "KHÔNG TÌM THẤY"}')
    else:
        print(f'❌ Vẫn chưa thấy thư mục "{folder_name}" trong MyDrive.')
        print('Danh sách hiện tại:', os.listdir('/content/drive/MyDrive'))

# Thử chạy kiểm tra
setup_path('HOCSAU')
```

```python
# CELL 2: Khai báo path
PROJECT_DIR = Path('/content/drive/MyDrive/HOCSAU')
APP_DIR = PROJECT_DIR / 'app_chomeo'
APP_PATH = APP_DIR / 'app_chomeo.py'

# Đường dẫn tới các model cũ
CATDOG_MODEL_PATH = PROJECT_DIR / "output" / "main2" / "best_cnn_cats_dogs_main2.keras"
BREED_MODEL_PATH = PROJECT_DIR / 'output' / 'oxford_pet_mobilenetv2' / 'best_mobilenetv2_oxford_pet.keras'


print("PROJECT_DIR:", PROJECT_DIR, "->", PROJECT_DIR.exists())
print("APP_DIR:", APP_DIR, "->", APP_DIR.exists())
print("APP_PATH:", APP_PATH, "->", APP_PATH.exists())
print("CATDOG_MODEL_PATH:", CATDOG_MODEL_PATH, "->", CATDOG_MODEL_PATH.exists())
print("BREED_MODEL_PATH:", BREED_MODEL_PATH, "->", BREED_MODEL_PATH.exists())
```

```python
from tensorflow import keras

catdog_model = keras.models.load_model(CATDOG_MODEL_PATH)
breed_model = keras.models.load_model(BREED_MODEL_PATH)

print("CATDOG_MODEL_PATH:", CATDOG_MODEL_PATH)
print("CATDOG model name:", catdog_model.name)
print("CATDOG input shape:", catdog_model.input_shape)

print("\nBREED_MODEL_PATH:", BREED_MODEL_PATH)
print("BREED model name:", breed_model.name)
print("BREED input shape:", breed_model.input_shape)
```
